```python
from matplotlib import pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 2, 9, 4, 7]
#plt.bar(x,y)
plt.hist(y)
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Scatter plot")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"red")

plt.subplot(1,2,2)
plt.plot(x,y,"blue")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,"red")
axes2.plot(x,y,"blue")

plt.savefig("test_image.png")
```
